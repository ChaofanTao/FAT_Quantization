The figure shows the 1*16 1-D Fourier Tranform basis functions.
The computation of the spetrum of weight consider all elements in the spatial domain,
and the spetrum replects the density of weight on each frequency.

In this paper, we view weight tensor as image and perform 1-D Fourier Tranform.

```python
# Visualization of the frequency basis functions of 1-D Fourier Tranform 

import torch 
import torchvision
import matplotlib.pyplot as plt
import numpy as np

def decompose(N):
    sqrt = int(np.sqrt(N))
    for i in range(sqrt, 0, -1):
        if N % i == 0:
            return i, N // i
        
num=16  # data length N
height, width = decompose(num)
fa = torch.zeros(1,num,2)

plt.figure(figsize=(height*2, width*2))
for j in range(num):
    fa[:,j,:] = 1
    a = torch.ifft(fa, 1)[..., 0]
    a = (a-a.min()) /(a.max()-a.min())
    # bilinear upsampling to make visualization smooth
    a = torch.nn.Upsample((300,300), mode="bilinear")(a.unsqueeze(0).unsqueeze(0)).squeeze(0).squeeze(0)

    img = a.numpy() #torchvision.transforms.ToPILImage()(a)
    plt.subplot(height, width, j+1)
    plt.title("{}".format(j))
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap="Blues")
#     plt.axis('off')
    fa[:,j,:] = 0
```

The figure shows the 8*8 2-D Fourier Tranform basis functions for different frequencies,
which is conjugate symmetrical. 

2-D Fourier Tranform is usually used in image compression and denoising.
For example, high-frequency component in images are usually noise and can be filtered

```python
num=8  # data length N*N
fa = torch.zeros(1,num, num, 2)

plt.figure(figsize=(num, num))
for i in range(num):
    for j in range(num):
        fa[:,i,j,:] = 1
        a = torch.ifft(fa, 2)[..., 0]
        a = (a-a.min()) /(a.max()-a.min())
        # bilinear upsampling to make visualization smooth
        a = torch.nn.Upsample((300,300), mode="bilinear")(a.unsqueeze(0)).squeeze(0).squeeze(0)
        
        img = a.numpy() #torchvision.transforms.ToPILImage()(a)
        plt.subplot(num, num, i*num+j+1)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(img, cmap="Blues")
    #     plt.axis('off')
        fa[:,i,j,:] = 0
```
